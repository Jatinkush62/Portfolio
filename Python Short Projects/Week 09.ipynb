# Week 9 Portfolio Exercise -  Visualizing the Titanic Disaster




### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

### Step 2. Import the dataset and Assign it to a variable titanic 

```python

titanic = pd.read_csv('titanic.csv')

titanic.head()
```

### Step 3. Set PassengerId as the index 

```python
titanic.set_index('PassengerId')
```

### Step 4. Create a pie chart presenting the male/female proportion

```python
sexList=titanic['Sex'].tolist()
labels=set(titanic['Sex'])
sexCounts=[sexList.count(i) for i in labels]
plt.axis("equal")
colors=['orange','teal']
plt.pie(sexCounts, labels=labels, autopct='%1.1f%%',colors=colors)
plt.legend(loc='lower right')
plt.title('Titanic Sex ratio')
plt.show()
```

### Step 5. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
sns.scatterplot(x='Age', y='Fare', hue='Sex', data=titanic)
plt.title("a scatterplot with the Fare payed and the Age, differ the plot color by gender")
plt.show()
```

### Step 6. How many people survived?

```python
print("1 is for survived:-")
titanic['Survived'].value_counts()
```

### Step 7. Create a histogram with the Fare payed

```python
plt.hist(titanic['Fare'], bins=10, color='blue', edgecolor='black')
plt.xlabel("Fare payed")
plt.ylabel("Frequency")
plt.title("Histogram with the Fare payed")
plt.show()
```

## Create a heatgraph with Sex, Age and Fare.

```python
titanic2 = ['Sex','Age','Fare']
subset = titanic[titanic2]
correlation_matrix = subset.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Age, Fare, and Passenger Class')
plt.show()
```
